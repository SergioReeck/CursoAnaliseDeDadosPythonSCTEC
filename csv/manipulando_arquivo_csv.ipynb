# Manipulando CSV com Módulo CSV 

```python
import csv

with open("funcionarios.csv", "w", newline="", encoding="utf-8") as arquivo:
    escrita = csv.writer(arquivo)

    escrita.writerow(["Nome", "Cargo", "Salário"])

    escrita.writerow(["Ana", "Analista de Dados", 4000.00])
    escrita.writerow(["Carlos", "Desenvolvedor de Software", 3500.00])
    escrita.writerow(["Fernanda", "Analista de Negócios", 5000.00])
```

```python
with open("funcionarios.csv", "r", encoding="utf-8") as arquivo:
    leitura = csv.reader(arquivo)

    for linha in leitura:
        print(linha)
```

# CSV com DictReader

```python
with open("filmes.csv", "r", encoding="utf-8") as arquivo:
    leitor = csv.DictReader(arquivo)

    for filme in leitor:
        print(filme)
```

```python
with open("filmes.csv", "r", encoding="utf-8") as arquivo:
    leitor = csv.DictReader(arquivo)

    for filme in leitor:
        print(f"Título : {filme['Título']}")
        print(f"Ano    : {filme['Ano']}")
        print(f"Gênero : {filme['Gênero']}")
        print("-" * 30)
```
