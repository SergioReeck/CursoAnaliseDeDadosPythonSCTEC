## Exemplo - Aluguel de Carro

Escreva um programa que pergunte a quantidade de km percorridos por um carro alugado pelo usuário, assim como a quantidade de dias pelos quais o carro foi alugado. Calcule o preço, sabendo que o carro custa 60,00 reais por dia e 0,15 centavos por km rodado.

```python
# Entrada de dados
qtde_km = int(input("Informe a quantidade de km percorridos: "))
qtde_dias = int(input("Informe a quantidade de dias em que o carro foi alugado: "))

# Cálculos
valor_diaria = qtde_dias * 60.00
valor_km = qtde_km * 0.15
total = valor_diaria + valor_km

# Saída de dados
print(f"O valor total a pagar é de R${total:.2f}")
```

### Exercício 2

Escreva uma expressão que será utilizada para decidir se um aluno foi ou não aprovado. Para ser aprovado, a média final do aluno deve ser maior ou igual a 7. Considere que o aluno cursa apenas três matérias, e que a nota de cada matéria está armazenada nas seguintes variávies:

- materia_1
- materia_2
- materia_3

Utilize como aprovação: True ou False

```python
materia_1 = float(input("Informe a nota da matéria 1: "))
materia_2 = float(input("Informe a nota da matéria 2: "))
materia_3 = float(input("Informe a nota da matéria 3: "))

media_final = (materia_1 + materia_2 + materia_3) / 3
aprovado = media_final >= 7
print(f"O aluno foi aprovado? {aprovado}, com média {media_final:.2f}")
```

## Exemplo - Frete Grátis

Você está desenvolvendo o sistema de uma loja online. O cliente terá direito ao "frete grátis" se atender a pelo menos uma das seguintes regras de negócio. **Regra 1:** Ele precisa ser um cliente "Premium" e o valor da compra deve ser maior ou igual a 100,00 reais. **Regra 2:** Ele pode ser um cliente comum, mas o valor da compra precisa ser maior que 250,00 reais e os produtos da compra não podem estar em promoção.

Dica: Utilize "input()" para comparar cenários. Variáveis do sistema:

- status_cliente (texto: "Premium" ou "Normal")
- valor_compra (número decimal)
- produto_promocao (booleano: True ou False)

```python
status_cliente = input("Status do cliente? ")
valor_compra = float(input("Qual o valor da compra?"))
produto_promocao = (input("Produto em promoção? (Sim ou Não)"))

frete_gratis = (
    (status_cliente == "Premium" and valor_compra >= 100.00) # Regra 1
     or
    (status_cliente == "Normal" and valor_compra > 250.00 and produto_promocao == "Não") # Regra 2
)

print(f"Frete grátis? {frete_gratis}")
```
