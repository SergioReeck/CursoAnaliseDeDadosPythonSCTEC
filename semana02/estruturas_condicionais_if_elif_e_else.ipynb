# Estruturas Condicionais if, elif e else

## Exemplo - Média com Estrutura Condicional

```python
materia_1 = float(input("Informe a nota da matéria 1: "))
materia_2 = float(input("Informe a nota da matéria 2: "))
materia_3 = float(input("Informe a nota da matéria 3: "))

media_final = (materia_1 + materia_2 + materia_3) / 3

if media_final >= 7:
    print(f"Aprovado com média {media_final:.2f}")
else:
    print(f"Reprovado com média {media_final:.2f}")

```

### Exercício 1

Escreva um programa que pergunte a velocidade do carro de um usuário. Caso ultrapasse 80 km/h, exiba um mensagem dizendo que o usuário foi multado. Nesse caso, exiba a velocidade, além do valor da multa, cobrando 5,00 reais por km acima de 80 km/h.




```python
velocidade = int(input("Informe a velocidade do carro: "))

if velocidade > 80:
    multa = (velocidade - 80) * 5
    print(f"Você foi multado! Sua velocidade foi de {velocidade} km\\h e o valor da multa é de R$ {multa:.2f}.")
```

# Estruturas Aninhadas

## Exemplo - Conta de Telefone com Três Faixas de Preço

```python
minutos = int(input("Informe a quantidade de minutos utilizados este mês: "))

if minutos < 200:
  preco = 0.20
else:
  if minutos < 400:
    preco = 0.18
  else:
    preco = 0.15

print(f"O valor a pagar é de R$ {preco * minutos:.2f}")
```

### Exercício 1

Escreva um programa que leia dois números e pergunte qual operação você deseja realizar. Você deve poder calcular soma, subtração, multiplicação e divisão. Exiba o resultado da operação solicitada.

```python
# Entrada de Dados

num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Menu de Operações

operacao = input(
    "Escolha uma operação:\n" \
    "+ - Adição\n" \
    "- - Subtração\n" \
    "* - Multiplicação\n" \
    "/ - Divisão\n" \
    "Digite a operação desejada: "
    }
        
# Processamento e Saída de Dados

if operacao == "+":
    resultado = num1 + num2
    print(f"Resultado: {resultado}")
elif operacao == "-":
    resultado = num1 - num2
    print(f"Resultado: {resultado}")
elif operacao == "*":
    resultado = num1 * num2
    print(f"Resultado: {resultado}")
elif operacao == "/":
    if num2 != 0:
        resultado = num1 / num2
        print(f"Resultado: {resultado}")
    else:
        print("Erro: Divisão por zero não é permitida.")
else:
    print("Operação inválida.")
```

### Execício 2

Escreva um programa que pergunte o salário do funcionário e calcule o valor do aumento. Para salários superiores à 2.000,00 reais, calcule um aumento de 10 por cento. Para salários inferiores ou iguais, um aumento de 15 por cento. Exiba o valor do salário atual, o percentual de aumento, o valor de aumento, e o salário após aumento.


```python
salario_atual = float(input("Informe o salário atual: R$ "))

if salario_atual > 2000:
    percentual_aumento = 0.10
else:
    percentual_aumento = 0.15 

valor_aumento = salario_atual * percentual_aumento
novo_salario = salario_atual + valor_aumento

print(f"Salário atual: R$ {salario_atual:.2f}")
print(f"Percentual de aumento: {percentual_aumento * 100:.0f}%")
print(f"Valor do aumento: R$ {valor_aumento:.2f}")
print(f"Novo salário: R$ {novo_salario:.2f}")
```

# Repetições

## While


```python
contador = 1

while contador <= 10:
    print(contador)
    contador = contador + 1
```

```python
numero = 0

while numero != 99:
    numero = int(input("Digite um número (99 para sair): "))
print("Programa encerrado.")
```

```python
while True:
    numero = int(input("Digite um número inteiro (ou 99 para sair): "))
    
    if numero == 99:
        break
    print("Programa encerrado.")
```

## For

```python
for numero in range(1, 11):
    print(numero)
```

```python
for nome in ["Alice", "Bob", "Charlie", "David", "Eva"]:
    print(nome)
```
