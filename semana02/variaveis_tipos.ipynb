## Exemplo - Cálculo de Aumento Salarial

```python
# Dados do colaborador
salario_atual = 2000.00
percentual_aumento = 5

# Cálculos
valor_aumento = salario_atual * (percentual_aumento / 100)
novo_salario = salario_atual + valor_aumento

# Novo Salário
print(f'O novo salário do colaborador é de: R$ {novo_salario}')
```

## Prática - Dividindo a Conta do Restaurante
