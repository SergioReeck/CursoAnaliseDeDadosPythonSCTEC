## Dicionários

```python
precos = {"arroz": 15.90, "feijao": 5.80, "leite": 5.90}

print(precos)

print(f"O preço do leite é: R$ {precos['leite']:.2f}")
```

```python
precos["leite"] = 5.99

print(f"O preço do leite atualizado é: R$ {precos['leite']:.2f}")
```

```python
if "arroz" in precos:
    print("O arroz está disponível.")
else:
    print("O arroz não está disponível.")
```

```python
print(precos.keys())

# dict_keys = (['arroz', 'feijao', 'leite'])
```
