## Listas

```python
# Criando lista de cidades

cidades = ["Florianópolis", "Criciúma", "Blumenau", "Joinville", "Timbó"]

print(cidades[2])  # Acessando o elemento na posição 2 da lista
```

### Adicionando Elementos na Lista

```python
#.append()

cidades.append("Jaraguá do Sul")  # Adicionando um elemento ao final da lista
cidades.append("Chapecó")  # Adicionando outro elemento ao final da lista
```

```python
print(cidades)  # Imprimindo a lista completa de cidades
```

```python
#.insert()

cidades.insert(1, "Balneário Camboriú")  # Inserindo um elemento na posição 1 da lista
print(cidades)  # Imprimindo a lista completa de cidades após a inserção
```

### Removendo Elementos da Lista

```python
#.remove()

cidades.remove("Timbó")  # Removendo o elemento "Timbó" da lista
print(cidades)  # Imprimindo a lista completa de cidades após a remoção
```

```python
#.pop()

removido = cidades.pop(3)  # Removendo o elemento na posição 3 da lista
print(f"Cidade removida: {removido}")  # Imprimindo o elemento
print(cidades)  # Imprimindo a lista completa de cidades após a remoção
```

```python
#.del()

del cidades[3]  # Removendo o elemento na posição 3 da lista
print(cidades)  # Imprimindo a lista completa de cidades após a remoção
```

```python
#len()

print(len(cidades))  # Imprimindo o comprimento da lista de cidades
```

### Fatiamento em Listas

```python
print(cidades[1:5])  # Acessando elementos da posição 1 até a posição 5 da lista
```

```python
print(cidades[:4])  # Acessando os elementos das posições 0 até 3 da lista
```

### Substituição de Valores em Listas

```python
print(cidades)  # Imprimindo a lista completa de cidades
```

```python
print(cidades[1:])  # Acessando os elementos da posição 1 até o final da lista
```

```python
cidades[1] = "Lages"  # Alterando o elemento na posição 1 da lista
print(cidades)  # Imprimindo a lista completa de cidades após a alteração
```
