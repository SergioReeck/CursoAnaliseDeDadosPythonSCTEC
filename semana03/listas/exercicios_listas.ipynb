### Exercício 1

Você foi ao supermercado e decidiu criar uma lista de compras em Python. Crie uma lista chamada compras contendo os seguintes produtos: Arroz, Feijão, Leite, Café, Açúcar. Depois, realize as atividades ao lado.

Atividades:

- Exiba a lista completa.
- Adicione o produto Macarrão utilizando append().
- Insira o produto Óleo na posição 2 utilizando insert().
- Remova o produto Leite utilizando remove().
- Remova o último produto da lista utilizando pop().
- Mostre a quantidade de produtos utilizando len().
- Exiba apenas os três primeiros produtos utilizando fatiamento.
- Substitua Café por Achocolatado.
- Utilize um for para exibir todos os produtos da lista.

```python
lista_mercado = ["arroz", "feijão", "leite", "café", "açúcar"]

print(lista_mercado)  # Imprimindo a lista completa de itens do mercado

lista_mercado.append("macarrão")  # Adicionando um novo item à lista

lista_mercado.insert(2, "óleo")  # Inserindo um item na posição 2 da lista

lista_mercado.remove("leite")  # Removendo o item "leite" da lista

lista_mercado.pop(-1)  # Removendo o último item da lista

print(len(lista_mercado))  # Imprimindo a lista completa de itens do mercado

print(lista_mercado[:3])  # Imprimindo os três primeiros itens da lista

lista_mercado[3] = "achocolatado"  # Alterando o item na posição 4 da lista

for item in lista_mercado:
    print(item)  # Imprimindo cada item da lista em uma linha separada
```
