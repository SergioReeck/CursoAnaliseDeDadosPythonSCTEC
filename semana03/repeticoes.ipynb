# WHILE


```python
soma = 0
contador = 1

while contador <=5:
    numero = int(input("Informe um número: ") or "0")
    soma += numero
    contador += 1

print(f"A soma dos números informados é: {soma}")
```

# FOR

```python
soma = 0

for nota in [1, 2, 3, 4]:
    valor = float(input(f"Informe um número {nota}: "))
    soma += valor

media = soma / 4
print(f"A média dos números informados é: {media}")
```
