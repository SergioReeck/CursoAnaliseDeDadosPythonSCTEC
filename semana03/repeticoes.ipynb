# WHILE


```python
soma = 0
contador = 1

while contador <=5:
    numero = int(input("Informe um número: "))
    soma += numero
    contador += 1

print(f"A soma dos números informados é: {soma}")
```

# FOR

```python
soma = 0
notas = [1, 2, 3, 4]

for nota in notas:
    valor = float(input(f"Informe a nota {nota}: "))
    soma += valor

media = soma / 4
print(f"A média das notas informadas é: {media}")
```
