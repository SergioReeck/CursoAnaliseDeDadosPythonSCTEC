```python
# Lista com números aleatórios

numeros = [17, 2, 4, 25, 18, 30, 41, 15, 8, 10, 53]

print("Lista original:", numeros)
print("---")

# Criação de listas vazias
pares = []
impares = []

# Percorrer a lista e encontrar valores

for numero in numeros:
    if numero % 2 == 0:
        pares.append(numero)
    else:
        impares.append(numero)

print("Números pares: ", pares)
print("---")
print("Números ímpares:", impares)
```
