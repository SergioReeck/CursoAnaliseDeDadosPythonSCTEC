```python
with open("alunos.txt", "w") as arquivo:
    arquivo.write("Ana\n")
    arquivo.write("Aline\n")
    arquivo.write("Fernanda")
```

```python
with open("alunos.txt", "r") as leitura:
    print(leitura.read())
```

```python
with open("alunos.txt", "a") as adicionar:
    adicionar.write("\nCarlos")
    adicionar.write("\nFelipe")
    
```

```python
with open("alunos.txt", "r+") as leitura_escrita:
    print(leitura_escrita.read())

    leitura_escrita.write("\nMarcos")
```

```python
with open("alunos.txt", "w+") as arquivo:
    arquivo.write("Ana\n")
    arquivo.write("Aline\n")

    arquivo.seek(3)

    print(arquivo.read())
```

```python
with open("alunos.txt", "r") as leitura:
    for linha in leitura:
        print(linha)
```

```python
contador = 0

with open("alunos.txt", "r") as leitura:
    for linha in leitura:
        contador += 1

print(contador)
```

# Criação de Arquivo e Inserção de Dados com Input

```python
with open("nomes.txt", "w") as arquivo:
    for i in range(5):
        nome = input("Nome: ")
        arquivo.write(nome + "\n")
```
