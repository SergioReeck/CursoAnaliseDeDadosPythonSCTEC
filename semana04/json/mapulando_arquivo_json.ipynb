# Exemplo 1

```python
# Importação de Bibliotecas

import json
```

```python
aluno = {
    "nome": "Ana",
    "idade": 25,
    "curso": "Análise de Dados",
}
```

```python
with open("aluno.json", "w", encoding="utf-8") as arquivo:
    json.dump(aluno, arquivo, indent=4, ensure_ascii=False)
```

```python
with open("aluno.json", "r", encoding="utf-8") as arquivo:
    aluno = json.load(arquivo)

    print(aluno)
    print(aluno["nome"])
```

# Exemplo 2

```python
import json
```

```python
alunos = [
    {"nome": "Ana", "nota": 8.5},
    {"nome": "Carlos", "nota": 9.2},
    {"nome": "Fernanda", "nota": 7.5}
]
```

```python
with open("alunos.json", "w", encoding="utf-8") as arquivo:
    json.dump(alunos, arquivo, indent=4, ensure_ascii=False)
```

```python
with open("alunos.json", "r", encoding="utf-8") as arquivo:
    alunos = json.load(arquivo)
```

```python
for aluno in alunos:
    print(aluno)
```

```python
for aluno in alunos:
    print(aluno["nome"], " - ", aluno["nota"])
```

# Exemplo 3

```python
aluno1 = {
    "nome": "Ana",
    "idade": 25,
    "disciplinas": [
        "Python",
        "SQL",
        "Power BI"
    ]}
```

```python
with open("aluno_lista.json", "w", encoding="utf-8") as arquivo:
    json.dump(aluno1, arquivo, indent=4, ensure_ascii=False)
```

```python
with open("aluno_lista.json", "r", encoding="utf-8") as arquivo:
    aluno_lido= json.load(arquivo)
```

```python
print(aluno_lido["nome"])
print("-" * 30)
print(aluno_lido["disciplinas"])
print("-" * 30)
print(aluno_lido["disciplinas"][0])
```

# Exemplo 4

```python
alunos_dict_list = [
    {
        "nome": "Ana",
        "idade": 22,
        "disciplinas": ["Python", "SQL", "Power BI"]
    },
    {
        "nome": "Carlos",
        "idade": 25,
        "disciplinas": ["Excel", "Python"]
    },
    {
        "nome": "Fernanda",
        "idade": 21,
        "disciplinas": ["Banco de Dados", "Estatística", "Machine Learning"]
    }
]

with open("alunos_dict_list.json", "w", encoding="utf-8") as arquivo:
    json.dump(alunos_dict_list, arquivo, indent=4, ensure_ascii=False)
```

```python
with open("alunos_dict_list.json", "r", encoding="utf-8") as arquivo:
    alunos_lidos = json.load(arquivo)

for aluno in alunos_lidos:
    print(f"Nome: {aluno['nome']}")
    print("Disciplinas:")

    for disciplina in aluno["disciplinas"]:
        print("-", disciplina)

    print()
```
